```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/scamhac/dataset_datahack_videos-dataset_datahack_videos.csv')
df.head(10)
```

```python
df.columns
```

```python

selected_column = df[['duration', 'view_count']]

selected_column
```

```python
# Convert duration to seconds
def convert_to_seconds(duration):
    duration = duration.replace('PT', '')
    minutes, seconds = duration.split('M')
    seconds = seconds.replace('S', '')
    return int(minutes)*60 + int(seconds)

df['duration'] = df['duration'].apply(convert_to_seconds)
# Define success based on views
def determine_success(views):
    if views >= 1000000:
        return 1
    else:
        return 0

df['success'] = df['views'].apply(determine_success)

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Define the features and target variable
features = ['duration']
target = 'success'

# Train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train_data[features], train_data[target])

# Make predictions on the testing set
y_pred = rf.predict(test_data[features])

# Convert predictions to binary based on threshold
y_pred_binary = [1 if y >= 0.5 else 0 for y in y_pred]

# Evaluate the model
mse = mean_squared_error(test_data[target], y_pred_binary)
r2 = r2_score(test_data[target], y_pred_binary)
print(f"Mean Squared Error: {mse}")
print(f"R^2: {r2}")
```
